```python
New_AAPL_df = AAPL_info_df.drop(columns=['Adj Close'])
```

```python
def add_commas(volume):
    return f"{volume:,.0f}" 
New_AAPL_df['Volume'] = New_AAPL_df['Volume'].apply(add_commas)
print(New_AAPL_df)
```

```python
New_AAPL_df
```

```python
New_AAPL_df[['Low', 'High', 'Open', 'Close']] = New_AAPL_df[['Low', 'High', 'Open', 'Close']].round(2).applymap("{:.2f}".format)
New_AAPL_df
```

```python
New_AAPL_df['Ticker'] = 'AAPL'
New_AAPL_df
```

```python
New_AAPL_df.info()
```

```python
#LOADING INTO POSTGRE SQL
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
#from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import DeclarativeBase

class Base(DeclarativeBase):
    pass
```

```python
class AAPL(Base):
    __tablename__ = 'New_AAPL_df'
    id = Column(Integer, primary_key=True)
    Date = Column(String(255))
    High = Column(String(255))
    Low = Column(String(255))
    Open = Column(String(255))
    Close = Column(String(255))
    Volume = Column(String(255))
    Ticker = Column(String)
```

```python
# Create Database Connection for Postgres
dbName = 'Projectstock'
userName = 'postgres'
password = 'Password'
engine = create_engine(f"postgresql://{userName}:{password}@localhost/{dbName}")
conn = engine.connect()
```

```python
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import Session # can be used to push data into the database or run queries from the database
session = Session(bind=engine)
```

```python
print(New_AAPL_df)
```

```python
# Convert DataFrame to dictionary
df_dict = New_AAPL_df.to_dict(orient='records')
```

```python
session.bulk_insert_mappings(AAPL, df_dict)
session.commit() 
```

```python
# Query to check the first few rows in the table
result = session.query(AAPL).limit(5).all()
for row in result:
    print(row.Date, row.High, row.Low, row.Open, row.Close, row.Volume, row.Ticker)
```
