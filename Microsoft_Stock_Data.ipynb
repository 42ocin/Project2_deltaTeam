```python
import pandas as pd
from pprint import pprint
import datetime as dt
```

```python
MicrosoftPath = "resources/MSFT.csv"
MicroStock = pd.read_csv(MicrosoftPath)
```

```python
MicroStock.head()
```

```python
#get the index datatype, column names
MicroStock.info()
```

```python
#put the columns into a list
MicroStockColumns = MicroStock.columns.tolist()
MicroStockColumns
```

```python
#get each row into a list
MicroStockValues = MicroStock.values.tolist()
MicroStockValues
```

```python
#adding columns to a list (did not end up being necessary)
new_columns = []
for col in MicroStockColumns:
    new_columns.append(col.split(","))
print(new_columns)
```

```python
#adding values to a list
MicroStockList = []
for stock in MicroStockValues:
    MicroStockList.append(stock)
MicroStockList
```

```python
#creating a data frame
MicroStockDF = pd.DataFrame(MicroStockList, columns = MicroStockColumns)
MicroStockDF
```

```python
#Dropping the adjusted close column
updatedMicroStockDF = MicroStockDF.drop('Adj Close', axis=1)
updatedMicroStockDF
```

```python
#round the values in the dataframe
roundedDF = updatedMicroStockDF.round(4)
roundedDF
```

```python
#add commas to volume to make it more readable
def add_commas(x):
    return "{:,}".format(x)
roundedDF['Volume'] = roundedDF['Volume'].apply(add_commas)
```

```python
#add a ticker column that can work as a primary key
roundedDF['Ticker'] = 'MSFT'
roundedDF
```

```python
#convert date to datetime
roundedDF['Date'] = pd.to_datetime(roundedDF['Date'])
roundedDF
```

```python
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float, DateTime
```

```python
from sqlalchemy.orm import DeclarativeBase
class Base(DeclarativeBase):
    pass
```

```python
class Microsoft(Base):
    __tablename__ = 'Tech_Stock_Data'
    Ticker = Column(String(255), primary_key=True)
    High = Column(Float)
    Low = Column(Float)
    Open = Column(Float)
    Close = Column(Float)
    Volume = Column(Float)
    Date = Column(DateTime)
```

```python
# Create Database Connection for Postgres
dbName = 'Tech_Stock_Data'
userName = 'postgres'
password = 'password'
engine = create_engine(f"postgresql://{userName}:{password}@localhost/{dbName}")
conn = engine.connect()
```

```python
from sqlalchemy.orm import Session 
session = Session(bind=engine)
```

```python
Base.metadata.create_all(engine)
```

```python
roundedDF.to_sql('Tech_Stock_Data', engine, if_exists='replace', index=False)
```
